# Exploratory Data Analysis(EDA)

```python
#To show amount spent by gender.
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
```

```python
#To show amount spent by by gender by a graph.
```

```python
ax = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).plot(
    kind='bar', x='Gender', y='Amount', color='red'
)
ax.set_title('Total Amount spent by Gender')
ax.set_xlabel('Gender')
ax.set_ylabel('Total Amount')
```

```python
#To show amount spent for each age group.
```

```python
df.groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
```

```python
#To count the number of males and females according to age group and show a graph.
```

```python
sns.countplot(data=df,x='Age Group',hue='Gender')
```

```python
#To show the label here in the graph.
```

```python
ax=sns.countplot(data=df,x='Age Group',hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#To show amount spent by each age group.
```

```python
ax=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).plot(kind='bar', x='Age Group', y='Amount', color='red')
ax.set_title('Total Amount by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Total Amount')
```

```python
#To count order given by 10 staates
```

```python
ax=df.groupby(['State'],as_index=False)['Orders'].count().sort_values(by='Orders',ascending=False).head(10).plot(kind='bar',x='State',y='Orders',color='green')
ax.set_title('Total order by State')
ax.set_xlabel('State')
ax.set_ylabel('Total Amount')

```

```python
#To show amount spent by each states.
```

```python
ax=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10).plot(kind='bar',x='State',y='Amount',color='skyblue')
ax.set_title('Total amount spent by State')
ax.set_xlabel('State')
ax.set_ylabel('Total Amount')
```

```python
#To analyze the gender wise selling for top four states.
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8,5))
sales_in_Uttarpradesh=df[df['State']=='Uttar Pradesh'].groupby('Gender')['Amount'].count().sort_values(ascending=False)
axs[0, 0].bar(sales_in_Uttarpradesh.index,sales_in_Uttarpradesh)
axs[0, 0].set_title('Gender Wise selling in Uttarpradesh')
axs[0, 0].tick_params(axis='x', rotation=45)
sales_in_Maharashtra=df[df['State']=='Maharashtra'].groupby('Gender')['Amount'].count().sort_values(ascending=False)
axs[0, 1].bar(sales_in_Maharashtra.index,sales_in_Maharashtra)
axs[0, 1].set_title('Gender Wise selling in Maharashtra')
axs[0, 1].tick_params(axis='x', rotation=45)
sales_in_Karnataka=df[df['State']=='Karnataka'].groupby('Gender')['Amount'].count().sort_values(ascending=False)
axs[1, 0].bar(sales_in_Karnataka.index,sales_in_Karnataka)
axs[1, 0].set_title('Gender Wise selling in Karnataka')
axs[1, 0].tick_params(axis='x', rotation=45)
sales_in_Delhi=df[df['State']=='Delhi'].groupby('Gender')['Amount'].count().sort_values(ascending=False)
axs[1, 1].bar(sales_in_Delhi.index,sales_in_Delhi)
axs[1, 1].set_title('Gender Wise selling in Delhi')
axs[1, 1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
#To analyze the age wise selling according to the states.
```

```python
fig, axs = plt.subplots(2, 2, figsize=(8,5))
sales_in_Uttarpradesh=df[df['State']=='Uttar Pradesh'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[0, 0].bar(sales_in_Uttarpradesh.index,sales_in_Uttarpradesh)
axs[0, 0].set_title('Age Wise selling in Uttarpradesh')
axs[0, 0].tick_params(axis='x', rotation=45)
sales_in_Maharashtra=df[df['State']=='Maharashtra'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[0, 1].bar(sales_in_Maharashtra.index,sales_in_Maharashtra)
axs[0, 1].set_title('Age Wise selling in Maharashtra')
axs[0, 1].tick_params(axis='x', rotation=45)
sales_in_Karnataka=df[df['State']=='Karnataka'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[1, 0].bar(sales_in_Karnataka.index,sales_in_Karnataka)
axs[1, 0].set_title('Age Wise selling in Karnataka')
axs[1, 0].tick_params(axis='x', rotation=45)
sales_in_Delhi=df[df['State']=='Delhi'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[1, 1].bar(sales_in_Delhi.index,sales_in_Delhi)
axs[1, 1].set_title('Age Wise selling in Delhi')
axs[1, 1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```

```python
#To show the amount spent for topmost 4 categories
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,5))
sales_of_Food=df[df['Product_Category']=='Food'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[0, 0].bar(sales_of_Food.index,sales_of_Food,color='green')
axs[0, 0].set_title('Age Wise selling of Food')
axs[0, 0].tick_params(axis='x', rotation=45)
sales_of_ElectronicsGadgets=df[df['Product_Category']=='Electronics & Gadgets'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[0, 1].bar(sales_of_ElectronicsGadgets.index,sales_of_ElectronicsGadgets,color='red')
axs[0, 1].set_title('Age Wise selling of Electronics & Gadgets')
axs[0, 1].tick_params(axis='x', rotation=45)
sales_of_clothingapparel=df[df['Product_Category']=='Clothing & Apparel'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[1, 0].bar(sales_of_clothingapparel.index,sales_of_clothingapparel)
axs[1, 0].set_title('Age Wise selling of clothing & apparel',color='blue')
axs[1, 0].tick_params(axis='x', rotation=45)
sales_of_FootwearShoes=df[df['Product_Category']=='Footwear & Shoes'].groupby('Age Group')['Amount'].sum().sort_values(ascending=False)
axs[1, 1].bar(sales_of_FootwearShoes.index,sales_of_FootwearShoes,color='pink')
axs[1, 1].set_title('Age Wise selling of Footwear & Shoes')
axs[1, 1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
```
